Собрать все статьи в один датасет. Название статьи, сслыка, том, год

```python
import pandas as pd
import os
import ast
import csv

volumes_data = pd.read_csv('urls.csv')  

publications = []

for file_name in os.listdir('titles/'): 
    if file_name.endswith('.txt'):
        volume = file_name.replace('.txt', '')
        with open(f'titles/{file_name}', 'r', encoding="utf-8") as file:
            for line in file:
                el = ast.literal_eval(line)
                publication_name, link = el[0], el[1]
                publications.append({'publication_name': publication_name, 'publication_url': link, 'tome': volume})

publications_df = pd.DataFrame(publications)

merged_df = publications_df.merge(volumes_data, on='tome')

merged_df.to_csv('dataset_complete.csv', index=False, sep=';', quoting=csv.QUOTE_NONNUMERIC, quotechar='"')
```

Ищем публикации в OpenAlex

```python
import pandas as pd
import requests
import time

def get_publication_info(title):
    base_url = "https://api.openalex.org/works"
    params = {
        "filter": f"title.search:{title}"
    }
    response = requests.get(base_url, params=params)
    if response.status_code == 200:
        results = response.json().get('results', [])
        if results:
            return results[0]
        else:
            return None
    else:
        print(f"Ошибка при запросе данных для публикации '{title}': {response.status_code}")
        return None

publications_df = pd.read_csv('dataset_complete.csv')

found_publications = []
not_found_publications = []

for _, row in publications_df.iterrows():
    publication_info = get_publication_info(row['publication_name'])
    if publication_info:
        authors = [author['author']['display_name'] for author in publication_info.get('authorships', [])]
        publication_authors = {
            'author' : authors,
            'year' : row['year'],
            'title' : row['publication_name']
        }
        found_publications.append(publication_authors)
    else:
        not_found_publications.append(row['publication_name'])
    time.sleep(0.1)

found_publications_df = pd.DataFrame(found_publications)

found_publications_df.to_csv('found_publications.csv', index=False)

total_articles = len(publications_df)
not_found_ratio = len(not_found_publications) / total_articles

print(f"Количество ненайденных статей: {len(not_found_publications)}")
print(f"Доля ненайденных статей: {not_found_ratio:.2%}")

not_found_publications_df = pd.DataFrame(not_found_publications)
not_found_publications_df.to_csv('not_found_publications.csv', index=False)
```

Переводим found_publications в json построчно и сохраняем отдельно

```python
import pandas as pd
import os
import json

df = pd.read_csv('found_publications.csv', sep=';')

df['id'] = df['id'].str.split('/').str[-1]

folder_path = 'json_files'
os.makedirs(folder_path, exist_ok=True)

for index, row in df.iterrows():
    row_dict = row.to_dict()
    
    file_name = f"{row['id']}.json"
    
    file_path = os.path.join(folder_path, file_name)
    
    with open(file_path, 'w') as json_file:
        json.dump(row_dict, json_file, indent=4)

    print(f"Saved: {file_path}")
```

Скачиваем историю публикаций найденных авторов через id и сохраняем как json

```python
import requests
import json
import os
import time

with open("not_downloaded_authors.json", "r", encoding="utf-8") as f:
    authors = json.load(f)

output_folder = "authors_publications"

def get_publications_by_author(author_id):
    url = f"https://api.openalex.org/works?filter=author.id:{author_id}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Ошибка при запросе публикаций для автора {author_id}")
        return None
    time.sleep(0.1)

for author_url, author_name in authors.items():
    author_id = author_url.split('/')[-1]

    publications = get_publications_by_author(author_id)

    if publications:
        file_name = f"{author_id}_publications.json"
        file_path = os.path.join(output_folder, file_name)
        
        with open(file_path, "w", encoding="utf-8") as json_file:
            json.dump(publications, json_file, indent=4, ensure_ascii=False)
        
        # print(f"Публикации автора {author_name} сохранены в файл {file_path}")
```

Создаем временный файл с id авторов и оставляем только те что еще не нашлись

```python
import json
import os

# Чтение JSON-файла с ID авторов
with open("authors.json", "r", encoding="utf-8") as f:
    authors = json.load(f)

downloads_folder = "authors_publications"

downloaded_ids = [file_name.split('_')[0] for file_name in os.listdir(downloads_folder) if file_name.endswith('_publications.json')]

not_downloaded_authors = {author_id: name for author_id, name in authors.items() if author_id.split('/')[-1] not in downloaded_ids}

with open("not_downloaded_authors.json", "w", encoding="utf-8") as json_file:
    json.dump(not_downloaded_authors, json_file, indent=4, ensure_ascii=False)

print(f"Создан файл not_downloaded_authors.json с авторами, публикации которых не были скачаны.")
```
